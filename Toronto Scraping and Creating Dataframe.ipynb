# Scraping Data from the Wikipedia page

First, we import the necessary libraries

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

print('Libraries imported.')
```

```python
 from bs4 import BeautifulSoup # To scrape the table off wikipedia
```

Now, we find the relevant table from the page source code and use BeautifulSoup to to scrape the table.

```python
wiki_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(wiki_url, 'lxml')
cells= soup.find('table', class_ ="wikitable sortable")
print(cells.prettify())
```

```python
for rows in cells.find_all('tr'):
    print(rows.text)
```

```python
for tr in cells.find_all('tr'):
    data = (tr.find_all('td'))
    row1 = [i.text for i in data]
    print(row1)
```

## Creating Dataframe Using pandas

```python
row = []
for  tr in cells.find_all('tr'):
    data = tr.find_all('td')
    row.append([i.text.strip() for i in data])
labels = cells.find_all('th')
labels = [c.text for c in labels]
labels = [i.strip() for i in labels] ### removing '\n' from the column labels
print(labels)
df = pd.DataFrame( data  = row , columns = labels)
df.head()
```

```python
df1 = df.drop([0])   # remove the first empty row
df1.head()
```

```python
df1 = df1.reset_index(drop = True)
df1.head()
```

```python
df1=df1[df1.Borough != 'Not assigned']
df1.head()
```

```python
df1.columns
```

```python
df1=df1.groupby(['Postal Code','Borough'])['Neighborhood'].apply(','.join).reset_index()
df1
```
